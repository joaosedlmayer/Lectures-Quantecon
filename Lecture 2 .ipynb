## Blocos de Código e Indentação em Python

Em Python, um bloco de código é um conjunto de linhas de código que são executadas juntas. A extensão de um bloco de código é determinada pela sua indentação.

### Execução de Blocos de Código em Python

Em Python, as duas linhas indentadas são executadas `ts_length` vezes antes de continuar.

Essas duas linhas são chamadas de um "bloco de código", pois compreendem o "bloco" de código sobre o qual estamos iterando.

Ao contrário da maioria das outras linguagens, Python determina a extensão de um bloco de código apenas pela sua indentação.

No nosso programa, a indentação diminui após a linha `ϵ_values.append(e)`, indicando para Python que esta linha marca o limite inferior do bloco de código.

Mais informações sobre indentação serão discutidas posteriormente. Por agora, vamos ver outro exemplo de um loop "for".


````python
animals = ['dog', 'cat', 'bird']
for animal in animals:
    print("The plural of " + animal + " is " + animal + "s")
````

## Loop While

O loop `for`é o mais utilizado, mas o `While` também é utilizado. Por exemplo:


````python
ts_length = 100
ϵ_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    ϵ_values.append(e)
    i = i + 1
plt.plot(ϵ_values)
plt.show()
````


O loop "while" continuará executando o bloco de código delimitado pela indentação até que a condição (i < ts_length) seja satisfeita.

Neste caso, o programa continuará adicionando valores à lista `ϵ_values` até que `i` seja igual a `ts_length`:


````python
i == ts_length #the ending condition for the while loop
````

## Aplicação: Gráfico de Saldo de uma Conta Bancária ao Longo do Tempo

Nesta aplicação, nós plotamos o saldo de uma conta bancária ao longo do tempo.

Não há saques ao longo do período de tempo, cuja última data é denotada por \( T \).

O saldo inicial é \( B_0 \) e a taxa de juros é \( r \).

O saldo é atualizado do período \( t \) até \( T \) de acordo com a fórmula \( B_t = B_{t-1} \times (1 + r) \).

No código abaixo, geramos e plotamos a sequência \( B_t \).

Em vez de usar uma lista Python para armazenar essa sequência, usaremos um array NumPy.


````python
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty NumPy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
````



A declaração `b = np.empty(T+1)` aloca espaço na memória para T+1 números em um *floating point*.

Esses números são preenchidos pelo loop `for`.

Alocar memória no início é mais eficiente do que usar uma lista Python e o método `append()`, já que este último deve solicitar repetidamente espaço de armazenamento ao sistema operacional.



## Exercícios:

Temos que simular e plotar a time series correlacionada:
$$ x_{t+1} = \alpha x_t + \epsilon_{t+1} \ where \ x_0 =0 \ and \ t=0,\cdots,T$$

Com a sequência de choques dada por $\{\epsilon_t\}$ sendo IID e segue uma normal padrão.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
α = 0.9 #valor do alpha
T = 200 #tempo
x = np.empty(T+1) #estamos criando um array para alocar todo x_t
x[0] = 0 #valor inicial do x_t

for t in range(T):
    x[t+1] = α * x[t] + np.random.randn()

plt.plot(x)
plt.show()
````

Note que: Aqui está a explicação passo a passo:

α = 0.9: Define uma variável chamada α com o valor 0.9. Esta variável será usada em um cálculo posterior.

T = 200: Define uma variável chamada T com o valor 200. Esta variável representa o número de períodos de tempo que desejamos modelar.

x = np.empty(T+1): Cria um array NumPy vazio chamado x com T+1 elementos. O uso de np.empty() aloca espaço na memória para o array, mas não inicializa os valores. O primeiro elemento do array (índice 0) é definido posteriormente como 0.

x[0] = 0: Define o primeiro elemento do array x como 0. Este é o valor inicial para o modelo.

O loop for seguinte é executado para cada valor de t de 0 a T-1 (ou seja, para 0, 1, 2, ..., 199):

x[t+1] = α * x[t] + np.random.randn(): Atualiza o próximo elemento do array x com base no valor atual e em um termo aleatório. Isso simula a evolução de uma série temporal onde o valor em cada período de tempo depende do valor anterior (multiplicado por α) e de um termo de erro aleatório (gerado por np.random.randn()).
plt.plot(x): Plota o array x usando a biblioteca Matplotlib. Isso cria um gráfico que mostra a evolução da série temporal ao longo do tempo.

plt.show(): Mostra o gráfico na saída do Jupyter Notebook.

No geral, este código gera uma série temporal simulada onde cada valor depende do valor anterior, com a adição de ruído aleatório. O resultado é um gráfico que mostra a evolução da série temporal ao longo do tempo.

### Exercício 3

Queremos plotar a seguinte times series:

$$x_{t+1} = \alpha |x_t| + \epsilon_{t+1} \ where \ x_0 = 0 \ and \ t=0,1,\cdots, T$$
Usando $T=200$ e $\alpha = 0,9$

````python
import numpy as np
import matplotlib.pyplot as plt

T = 200
α = 0.9
x = np.empty(T + 1)
x[0] = 0

for t in range(T):
    x[t + 1] = α * np.abs(x[t]) + np.random.randn()

plt.plot(x)
plt.xlabel('Tempo')
plt.ylabel('Valor de x')
plt.title(f'Série Temporal com α = {α}')
plt.show()
````

Agora queremos fazer o mesmo porém sem utilizar a função `abs`

````python
α = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    if x[t] < 0:
        abs_x = - x[t]
    else:
        abs_x = x[t]
    x[t+1] = α * abs_x + np.random.randn()

plt.plot(x)
plt.xlabel('Tempo')
plt.ylabel('Valor de x')
plt.title(f'Série Temporal com α = {α}')
plt.show()
````
